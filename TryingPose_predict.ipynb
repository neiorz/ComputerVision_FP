```python
from ultralytics import YOLO
import cv2
import os

# Load a pretrained YOLOv8n-pose model
model = YOLO('yolov8n-pose.pt')

# Source video path
source_video = "/content/drive/MyDrive/NTI-ComputerVision/YOLOv8_HumanPose/videos/3943396-hd_1920_1080_30fps.mp4"

# Perform inference on the video
results = model.predict(source_video, save=True, conf=0.25, iou=0.7, show_labels=False, show_conf=False, show_boxes=True, project='runs/pose/predict', name='yolo_predict_output')

print(f"Prediction complete. Results saved to: {model.predictor.save_dir}")
```
